**DECISION** **TREE** **CLASSIFIER**

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier()
from sklearn.tree import export_graphviz,plot_tree
```

```python
cncrdf = pd.read_csv("/content/cancer patient data sets.csv")
cncrdf
```

```python
feature_col = ["chronic Lung Disease","Smoking","Chest Pain","Dry Cough"]
features = cncrdf[feature_col]  #features
target = cncrdf.Level   #target variables
```

```python
model1.fit(features,target)
```

```python
model1.predict([[2,3,2,3]])
```

```python
CLD = int(input("Enter Chronic Lung Disease:"))
Smk = int(input("Enter Smoking: "))
CP  = int(input("Enter Chest Pain: "))
DC = int(input("Enter Dry Cough: "))
pred = model1.predict([[CLD,Smk,CP,DC]])
if pred== "Low":
  print("You dont have Cancer")
else:
  print("You have Cancer")
```

```python
plot_tree(model1)
```

**RANDOM** **FOREST**

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=10)
```

```python
cncrdf = pd.read_csv("/content/cancer patient data sets.csv")
cncrdf
```

```python
feature_col = ["chronic Lung Disease","Smoking","Chest Pain","Dry Cough"]
features1 = cncrdf[feature_col]  #features
target1 = cncrdf.Level   #target variables
```

```python
model2.fit(features1,target1)
```

```python
model2.predict([[2,2,2,7]])
#model2.predict([[6,8,7,2]])
```

```python
CLD = int(input("Enter Chronic Lung Disease:"))
Smk = int(input("Enter Smoking: "))
CP  = int(input("Enter Chest Pain: "))
DC = int(input("Enter Dry Cough: "))
pred1 = model2.predict([[CLD,Smk,CP,DC]])
if pred1== "Low":
  print("You dont have Cancer")
else:
  print("You have Cancer")
```

```python
import os
output_dir="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
model2.estimators_
```

```python
for i, tree in enumerate(model2.estimators_):
  tree_dot_file = os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file = os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["chronic Lung Disease","Smoking","Chest Pain","Dry Cough"],
                    class_names=[str(cls) for cls in model2.classes_], filled=True, rounded=True)
```

```python
command = f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```
